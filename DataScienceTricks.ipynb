```python
# this is how we create data frame usin dictionary
dict1 = {
    'first': np.random.randint(1,10,9),
    'second': np.random.randint(1,10,9),
    'last': np.random.uniform(0,10,9)
}
```

```python
data=pd.DataFrame(dict1)

# this is groupby first variable keeping last and second column, group is by sum, and then the values are soreted based on second
# this is qual to stata collapse
data.groupby('first')[['last','second']].sum().sort_values('second', ascending=False)
```

```python
# sorting values first and grouping
ok=data.sort_values(by='first', ascending=True).groupby([data['second'],data['first']])
# ignore below
for key, values in ok:
    print(ok.get_group(key), "\n")
```
